- 20년 국내인구 총이동량 증가

- 베이비붐 세대 은퇴

    - 안정적 정착 지원

- 코로나 19 장기화로 인한 사회/경제적 여파
- 농촌에 대한 관심도
- 도시주택 가격 상승
---
- 정책
    - 체계적인 귀농/귀촌 준비 지원
